```python
# bài 1.kiem tra so nguyen to
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True 
# bài 2.kiem tra so hoan hao
def is_perfect(n):
    if n < 2:
        return False
    sum_of_divisors = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            sum_of_divisors += i
            if i != n // i:
                sum_of_divisors += n // i
    return sum_of_divisors == n
# bài 3.kiem 
```
