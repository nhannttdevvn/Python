### **Bài 2. Viết test cho hàm kiểm tra số nguyên tố:**

```python
def is_prime(n):
    if n <= 1: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# TODO: Viết test cho các số: 1, 2, 3, 4, 17, 18, 19
class TestIsPrime(unittest.TestCase):
    def test_is_prime(self):
        self.assertFalse(is_prime(1))
        self.assertTrue(is_prime(2))
        self.assertTrue(is_prime(3))
        self.assertFalse(is_prime(4))
        self.assertTrue(is_prime(17))
        self.assertFalse(is_prime(18))
        self.assertTrue(is_prime(19))

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 3. Viết test cho lớp `User`:**

```python
class User:
    def __init__(self, username):
        self.username = username

    def is_admin(self):
        return self.username == "admin"

# TODO: Viết test cho các username: 'admin', 'guest', 'user123'
class TestUser(unittest.TestCase):
    def test_is_admin(self):
        self.assertTrue(User('admin').is_admin())
        self.assertFalse(User('guest').is_admin())
        self.assertFalse(User('user123').is_admin())

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 4. Mock hàm gửi email:**

```python
from unittest.mock import patch

def send_welcome_email(email):
    print(f"Sending email to {email}")

# TODO: Mock print() và xác nhận nội dung gọi đúng
class TestSendWelcomeEmail(unittest.TestCase):
    @patch('builtins.print')
    def test_send_welcome_email(self, mock_print):
        send_welcome_email("test@example.com")
        mock_print.assert_called_with("Sending email to test@example.com")

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 5. Mock gọi API với requests.get:**

```python
import requests
from unittest.mock import patch

def fetch_user():
    response = requests.get("https://api.example.com/user")
    return response.json()

# TODO: Dùng unittest.mock để giả lập kết quả requests.get()
class TestFetchUser(unittest.TestCase):
    @patch('requests.get')
    def test_fetch_user(self, mock_get):
        mock_get.return_value.json.return_value = {"id": 1, "name": "John Doe"}
        result = fetch_user()
        self.assertEqual(result, {"id": 1, "name": "John Doe"})
        mock_get.assert_called_with("https://api.example.com/user")

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 6. Mock ngày hiện tại để test cuối tuần:**

```python
from datetime import datetime
from unittest.mock import patch, Mock

def is_weekend():
    today = datetime.now().weekday()
    return today >= 5

# TODO: Mock datetime.now() để test với thứ 6 (4) và thứ 7 (5)
class TestIsWeekend(unittest.TestCase):
    @patch('__main__.datetime')
    def test_is_weekend(self, mock_dt):
        mock_today = Mock()
        mock_dt.now.return_value = mock_today
        mock_today.weekday.return_value = 4
        self.assertFalse(is_weekend())
        mock_today.weekday.return_value = 5
        self.assertTrue(is_weekend())

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 7. Viết test cho hàm kiểm tra mật khẩu:**

```python
def is_strong(password):
    return len(password) >= 8 and any(c.isdigit() for c in password)

# TODO: Test với các password yếu và mạnh, kiểm tra đúng/false
class TestIsStrong(unittest.TestCase):
    def test_is_strong(self):
        self.assertFalse(is_strong("weak"))           
        self.assertFalse(is_strong("password"))    
        self.assertTrue(is_strong("StrongPass123"))   

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 8. Viết test cho hàm xử lý chuỗi:**

```python
def clean_input(s):
    return s.strip().lower().replace(" ", "_")

# TODO: Test với chuỗi có khoảng trắng, chữ in hoa, nhiều khoảng trắng
class TestCleanInput(unittest.TestCase):
    def test_clean_input(self):
        self.assertEqual(clean_input("  Hello World  "), "hello_world")
        self.assertEqual(clean_input("PYTHON"), "python")
        self.assertEqual(clean_input("A B C"), "a_b_c")

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

### **Bài 9. Viết test cho hàm chia số:**

```python
def safe_divide(a, b):
    if b == 0:
        return None
    return a / b

# TODO: Test các trường hợp: chia 10/2, 5/0, -4/2
class TestSafeDivide(unittest.TestCase):
    def test_safe_divide(self):
        self.assertEqual(safe_divide(10, 2), 5.0)
        self.assertIsNone(safe_divide(5, 0))
        self.assertEqual(safe_divide(-4, 2), -2.0)

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```
